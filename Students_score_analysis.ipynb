# Gender distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data=df, x='Gender', hue='Gender') 
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

# From the above chart we have analysed that the count of female is maximum

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb , annot = True)
plt.title("Relation between parent's education and student's score")
plt.figure(figsize= (5,5))
plt.show()
```

# From the above analysis we have concluded that the parents education has a great impact on students score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1 , annot = True)
plt.title("Relation between parent's marital status and student's score")
plt.figure(figsize= (5,5))
plt.show()
```

# From the above analysis we have concluded that the parents marital status has no or negligible impact on students score

# Distribution of ethnic groups

```python
groupA = df.loc[(df ['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df ['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df ['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df ['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df ['EthnicGroup'] == "group E")].count()

l = ['groupA' , 'groupB' , 'groupC' , 'groupD' , 'groupE']
mlist = [groupA["EthnicGroup"] , groupB["EthnicGroup"] , groupC["EthnicGroup"] , groupD["EthnicGroup"] , groupE["EthnicGroup"]]
plt.pie(mlist , labels=l , autopct = "%1.2f%%")
plt.title("Distribution of Ethnic groups")
plt.show()
```

# Through the pie chart we have distributed the ethnic groups
